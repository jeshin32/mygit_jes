```python
from selenium.webdriver import Chrome
driver = Chrome()
driver.get(url)
html_selenium = driver.page_source
```

```python
html_selenium
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_selenium, 'html.parser')
soup.select('div[class=card__content]')
```

```python
import lxml.html
root = lxml.html.fromstring(html_selenium)
root
links = root.cssselect('div.card__content')
links
```

```python
reviews = driver.find_elements_by_css_selector('div.card__content')
```

```python
for review in reviews:
    print(review.text)
```
